### Объединение таблиц `pd.DataFrame`

5.1 При помощи объединения таблиц, создайте `DataFrame`, состоящий из четырех столбцов: `id`, `name`, `user_id`, `rating`. Рецепты, на которые не оставлен ни один отзыв, должны отсутствовать в полученной таблице. Подтвердите правильность работы вашего кода, выбрав рецепт, не имеющий отзывов, и попытавшись найти строку, соответствующую этому рецепту, в полученном `DataFrame`.

```python
rating_table = recipes[['id', 'name']].merge(reviews[['user_id', 'rating', 'recipe_id', 'review']],
                                          left_on = 'id', right_on = 'recipe_id', how = 'inner')

rating_table.drop(columns = ['recipe_id', 'review'], inplace = True)
rating_table
# так как при объединении таблиц используется how = 'inner', то в итоговой таблице будут отсутствовать 
# рецепты, на которые отсутсвуют отзывы
```

```python
# проверка на наличие рецепта без отзыва в новой таблице, id рецепта взято из задания 4.2
rating_table[(rating_table.id == 47)]
```

5.2 При помощи объединения таблиц и группировок, создайте `DataFrame`, состоящий из трех столбцов: `recipe_id`, `name`, `review_count`, где столбец `review_count` содержит кол-во отзывов, оставленных на рецепт `recipe_id`. У рецептов, на которые не оставлен ни один отзыв, в столбце `review_count` должен быть указан 0. Подтвердите правильность работы вашего кода, выбрав рецепт, не имеющий отзывов, и найдя строку, соответствующую этому рецепту, в полученном `DataFrame`.

```python
review_table = recipes[['id', 'name']].merge(reviews[['recipe_id', 'review']],
                                             left_on = 'id', right_on = 'recipe_id', how = 'left')
review_table.drop('recipe_id', axis = 1, inplace = True)
review_table
```

```python
review_table_count = review_table.groupby('id')[['review']].count().reset_index()
review_table_count
```

```python
review_table.drop_duplicates(subset = 'id', inplace = True)
total_review_table = review_table[['id', 'name']].merge(review_table_count[['id', 'review']], on = 'id', how = 'left')
total_review_table.rename(columns = {'id':'recipe_id', 'review':'review_count'}, inplace = True)
total_review_table.sort_values(by = 'review_count')
```

```python
total_review_table[total_review_table.recipe_id == 375288]
```

5.3. Выясните, рецепты, добавленные в каком году, имеют наименьший средний рейтинг?

```python
reviews['date'] = pd.to_datetime(reviews['date'])
```

```python
mean_of_year = reviews.groupby(reviews.date.dt.year)[['rating']].mean().reset_index()
mean_of_year
```

```python
mean_of_year[mean_of_year['rating'] == mean_of_year['rating'].min()]
# вывод минимального рейтинга
```
